```python
test =pd.read_csv('../input/titanic/test.csv')
train = pd.read_csv('../input/titanic/train.csv')
```

```python
train['Embarked'] = train['Embarked'].fillna('S')
train['Fare'] =  train['Fare'].fillna(train['Fare'].median())
train['Age'] = train['Age'].fillna(train['Age'].median())
train['Cabin'] =  train['Cabin'].fillna('NULL')


train.loc[train['Age'] <=3, 'Age'] = 0
train.loc[(train['Age'] >3) & (train['Age'] <= 10), 'Age'] = 1
train.loc[(train['Age'] >10) & (train['Age'] <= 18), 'Age'] = 2
train.loc[(train['Age'] >18) & (train['Age'] <= 25), 'Age'] = 3
train.loc[(train['Age'] >25) & (train['Age'] <= 40), 'Age'] = 4
train.loc[(train['Age'] >40) & (train['Age'] <= 55), 'Age'] = 5
train.loc[(train['Age'] >55) & (train['Age'] <= 70), 'Age'] = 6
train.loc[train['Age'] >70, 'Age'] = 7
train.loc[train['Sex'] =='male', 'Sex'] = 0
train.loc[train['Sex'] =='female', 'Sex'] = 1
train.loc[train['Fare'] <=50, 'Fare'] = 0
train.loc[(train['Fare'] >50) & (train['Fare'] <=100 ), 'Fare'] = 1
train.loc[train['Fare'] >=100, 'Fare'] = 0
from sklearn.preprocessing import LabelEncoder
e = LabelEncoder()
train['embarked']=e.fit_transform(train['Embarked'])
input=train.drop(['Cabin','Embarked','Survived','Name','Ticket'],axis=1)
target=train['Survived']
input
```

```python
test['Embarked'] = test['Embarked'].fillna('S')
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Cabin'] =  test['Cabin'].fillna('NULL')


test.loc[test['Age'] <=3, 'Age'] = 0
test.loc[(test['Age'] >3) & (test['Age'] <= 10), 'Age'] = 1
test.loc[(test['Age'] >10) & (test['Age'] <= 18), 'Age'] = 2
test.loc[(test['Age'] >18) & (test['Age'] <= 25), 'Age'] = 3
test.loc[(test['Age'] >25) & (test['Age'] <= 40), 'Age'] = 4
test.loc[(test['Age'] >40) & (test['Age'] <= 55), 'Age'] = 5
test.loc[(test['Age'] >55) & (test['Age'] <= 70), 'Age'] = 6
test.loc[test['Age'] >70, 'Age'] = 7
test.loc[test['Sex'] =='male', 'Sex'] = 0
test.loc[test['Sex'] =='female', 'Sex'] = 1
test.loc[test['Fare'] <=50, 'Fare'] = 0
test.loc[(test['Fare'] >50) & (test['Fare'] <=100 ), 'Fare'] = 1
test.loc[test['Fare'] >=100, 'Fare'] = 0
from sklearn.preprocessing import LabelEncoder
b = LabelEncoder()
test['embarked']=b.fit_transform(test['Embarked'])
test_input=test.drop(['Cabin','Embarked','Name','Ticket'],axis=1)
test_input
```

```python
x_test = test_input.values
x_train=input.values
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=15)

model = model.fit(x_train,target)
y_pred = model.predict(x_test)


output = pd.DataFrame({'PassengerId': test_input.PassengerId, 'Survived': y_pred})
output.to_csv('my_submission1.csv', index=False)
print("Submission saved")
```
